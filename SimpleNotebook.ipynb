# What can you do with a Jupyter Notebook?

Basically anything you can do with Python!

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.widgets import Slider
```

```python
%matplotlib notebook
x = np.linspace(0, 1, 200)
y = np.sin(x * 2 * np.pi)

plt.plot(x, y)
plt.axes([0, 1, -1, 1])
plt.show()
```

```python
def my_polynomial(x):
    return 2*x**2 + x - 1
```

```python
for x in np.linspace(1,10,10):
    print('Input: {}\nOutput: {}\n'.format(x, my_polynomial(x)))
```
